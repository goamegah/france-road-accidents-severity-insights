### Téléchargement et chargement des données

```python
from bireport.data.data_loader import load_data
```

```python
df_2023 = load_data(2023)
df_2022 = load_data(2022)
df_2021 = load_data(2021)
df_2020 = load_data(2020)
df_2019 = load_data(2019)
```

### Nettoyage

```python
from  bireport.data.data_cleaning import clean_dataset
```

```python
df_2022_clean = clean_dataset(df_2022)
df_2023_clean = clean_dataset(df_2023)
```

### Fusionner les années

```python
from bireport.data.data_loader import merge_years
```

```python
df_all = merge_years([df_2022_clean, df_2023_clean])
```

### Annotations

```python
from bireport.data.data_annotation import annotate_columns
```
